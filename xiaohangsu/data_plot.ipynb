```python
filepath = 'plotData/'
borough = pd.read_csv(filepath + 'col14_borough.out', header=None, index_col=0, names=['Borough', 'Crime Total'])
borough
borough.plot(kind='bar', title='2006-2015 Total Crimes in Boroughs', rot=0)
```

```python
year = pd.read_csv(filepath + 'col2_year.out', header=None, index_col=0, names=['Year', 'Crime Total'])
year.plot(kind='bar', title='2006 - 2015 Total Crimes Each Year', rot=0)
```

```python
month = pd.read_csv(filepath + 'col2_month.out', header=None, index_col=0, names=['Month', 'Crime Total'])
month.plot(kind='bar', title='2006 - 2015 Total Crimes Each Month', rot=0)
```

```python
day = pd.read_csv(filepath + 'col2_day.out', header=None, index_col=0, names=['Day', 'Crime Total'])
day.plot(kind='bar', title='2006 - 2015 Total Crimes Each Day')
```

```python
hour = pd.read_csv(filepath + 'col3_hour.out', header=None, index_col=0, names=['Hour', 'Crime Total'])
hour.plot(kind='bar', title='Crimes Total at Each Hour', rot=0)
```

```python
dateGroup = pd.read_csv(filepath + 'col2_date_group.out', header=None, index_col=0, parse_dates=True, names=['date', 'Crimes'])
dateGroup.plot(figsize=(15, 6), title='Crimes Total in Date')
```

```python
timeGroup = pd.read_csv(filepath + 'col3_time_group.out', header=None, index_col=0, parse_dates=True, names=['Time', 'Crimes'])
timeGroup.index = timeGroup.index.time
timeGroup.plot(figsize=(15, 6), rot=0, title='Crimes Totals in Time')
```
