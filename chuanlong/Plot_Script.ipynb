```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset from CSV file
file_path = 'columns.csv'
df = pd.read_csv(file_path, header = None, sep = '\t', dtype = {1:str, 2:str})
```

```python
df.head()
```

```python
def analyse_text_attr(dataset, attribute, plot_type, figsize = (20, 10)):
    df = dataset[attribute]
    nan_count = df.isnull().sum()
    df = df.dropna()
    type_count = list(set(df))
    print('number of missing data =', nan_count)
    print('type of data =', type_count[:])
    df.value_counts().plot(kind = plot_type, figsize=figsize)
    plt.show()
```

```python
analyse_text_attr(df, 0, 'pie', (10, 10))
```

```python
analyse_text_attr(df, 1, 'bar', (20, 10))
```

```python
analyse_text_attr(df, 2, 'bar', (20, 10))
```

```python
analyse_text_attr(df, 3, 'bar', (20, 10))
```

```python
def analyse_coord_attr(dataset, x, y):
    coords = dataset[[x, y]]
    coords = coords.dropna()
    plt.figure(figsize=(20, 10))
    plt.scatter(coords[x].iloc[:], coords[y].iloc[:])
    plt.show()
```

```python
analyse_coord_attr(df, 4, 5)
```

```python
analyse_coord_attr(df, 6, 7)
```
