```python
def analyse_text_attr(dataset, attribute, plot_type):
    df = dataset[attribute]
    nan_count = df.isnull().sum()
    df = df.fillna('NaN')
    type_count = set(df)
    print('number of missing data =', nan_count)
    print('type of data =', type_count)
    df.value_counts().plot(kind = plot_type, figsize=(20,10))
    plt.show()
```

#### LOC_OF_OCCUR_DESC
* Data Type:Plain Text
* Specific location of occurrence in or around the premises(inside, opposite of, front of, rear of)
* Missing Data: 1127128 rows
* Invalid Data: ' '

```python
analyse_text_attr(df, 'LOC_OF_OCCUR_DESC', 'bar')
```

#### PREM_TYP_DESC
* Data Type:Plain Text
* Specific description of premises; grocery store, residence, street, etc.
* Missing Data: 33279 rows
* Invalid Data: 
* There are too many unique value, but some of them belong to the same thing. Should deal with it intuitively

```python
analyse_text_attr(df, 'PREM_TYP_DESC', 'bar')
```

#### PARKS_NM
* Data Type:Plain Text
* Name of NYC park, playground or greenspace of occurrence, if applicable (state parks are not included)
* Missing data: Most of them is missing, because the park name would exist if applicable

```python
analyse_text_attr(df, 'PARKS_NM', 'bar')
```

#### X_COORD_CD & Y_COORD_CD
* Data Type:Number
* X-coordinate for New York State Plane Coordinate System, Long Island Zone, NAD 83, units feet (FIPS 3104)
* Y-coordinate for New York State Plane Coordinate System, Long Island Zone, NAD 83, units feet (FIPS 3104)
* Missing Values: 188146
* Invalid Values: None(There is only one coordinate instead of two)
* Outline: None

```python
def analyse_coord_attr(dataset, x, y):
    coords = dataset[[x, y]]
    nan_coords = coords[coords[x].isnull() & coords[y].isnull()]
    nan_count = nan_coords.sum()
    print('number of missing data =', nan_count)
    plt.figure(figsize=(20, 10))
    plt.scatter(coords[x].iloc[:], coords[y].iloc[:])
    plt.show()
```

```python
analyse_coord_attr(df, 'X_COORD_CD', 'Y_COORD_CD')
```

```python
analyse_coord_attr(df, 'Latitude', 'Longitude')
```
